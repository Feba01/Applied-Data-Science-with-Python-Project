```python
# function to help clear up special characters in order to convert the datatype to float

def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
```

### 4) Box plots and Histograms to understand the distributions and outliers + outlier treatment.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Box plot to understand the distributions of the numeric variables

print ("Box plots")
numerical_variables = ['Income ', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
                  'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
                  'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
                  'Total_Children', 'Age', 'Total_Expenditure']
marketing_df[numerical_variables].plot(kind='box', subplots=True, layout=(4, 4), figsize=(15, 12), 
                        sharex=False, sharey=False)
plt.tight_layout()
plt.show()

# Histogram to understand the distributions of the numeric variables

print ("Histograms")
sns.set_style('darkgrid')
marketing_df[numerical_variables].hist(bins=20, figsize=(15, 12))
plt.tight_layout()
plt.show()
```

```python
# Outlier treatment
# Function to treat outliers using the IQR method

def treat_outliers(dataframe, column):
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    dataframe[column] = dataframe[column].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))

# To apply outlier treatment to each numerical variable except 'Age' and 'Total_Children'
for var in numerical_variables:
    if var not in ['Age', 'Total_Children']:
        treat_outliers(marketing_df, var)
        
# To check the effect of outlier treatment
marketing_df[numerical_variables].plot(kind='box', subplots=True, layout=(4, 4), figsize=(15, 12), sharex=False, sharey=False)
plt.tight_layout()
plt.show()
```

### 7) Hypotheses testing

```python
from scipy.stats import ttest_ind, kruskal

# Hypothesis : Older people prefer shopping in-store

older_customers = marketing_df[marketing_df['Age'] >= 50]
younger_customers = marketing_df[marketing_df['Age'] < 50]

t_stat, p_value = ttest_ind(older_customers['NumStorePurchases'], younger_customers['NumStorePurchases'])
print("Hypothesis 1:")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Null Hypothesis (H0):
There is no significant difference in preference for shopping in-store between older and younger individuals, i.e., age has no effect on the preference for shopping in-store.

Alternative Hypothesis (H1):
There is a significant difference in preference for shopping in-store between older and younger individuals. Specifically, older people are less likely to prefer shopping online and are more likely to shop in-store.

As the pvalue is very small (7.1188e-06) and the t-statistic which is 4.5007, is comparitively large. Hence, we reject the null hypothesis and accept the alternate hypotheses  which is that older people are less likely to prefer shopping online and are more likely to shop in-store.

```python
# Hypothesis: Customers with kids prefer shopping online

customers_with_kids = marketing_df[marketing_df['Total_Children'] > 0]
customers_without_kids = marketing_df[marketing_df['Total_Children'] == 0]

t_stat, p_value = ttest_ind(customers_with_kids['NumWebPurchases'], customers_without_kids['NumWebPurchases'])
print("\nHypothesis 2:")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Null Hypothesis (H0): There is no significant difference in the preference for shopping in-store between customers with kids and customers without kids.
Alternative Hypothesis (H1): Customers with kids have less time to visit a store and prefer to shop online compared to customers without kids.

The t-statistic value is -3.5447, indicating a relatively large negative value.
The p-value associated with this t-statistic is 0.00040, which is low and less than the typical statistic significance level of 0.05.
Hence, we reject the null hypothesis and accept the alternate hypotheses  which is that customers with kids have less time to visit a store and prefer to shop online compared to customers without kids.

```python
# Hypothesis: Other distribution channels cannibalize sales at the store

distribution_channels = ['NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
store_purchases = marketing_df['NumStorePurchases']
channel_purchases = marketing_df[distribution_channels]

h_stat, p_value = kruskal(*channel_purchases.values.T, store_purchases)
print("\nHypothesis 3:")
print("H-statistic:", h_stat)
print("p-value:", p_value)
```

Null Hypothesis (H0): Other distribution channels do not cannibalize sales at the store.
Alternate Hypothesis (H1): Other distribution channels may cannibalize sales at the store.

The test result shows a large H-statistic value of 1661.76 and a very small p-value of 0.0 (below the typical statistic significance level of 0.05).
Hence, we reject the null hypothesis and accept the alternate hypotheses  which is that the distribution channels other than the stores may cannibalize sales at the store.

```python
# Hypothesis: Does the US fare significantly better than the rest of the world in terms of total purchases?

us_customers = marketing_df[marketing_df['Country_US'] == 1]
rest_of_world_customers = marketing_df[marketing_df['Country_US'] == 0]

t_stat, p_value = ttest_ind(us_customers['Total_Expenditure'], rest_of_world_customers['Total_Expenditure'])
print("\nHypothesis 4:")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

Null Hypothesis (H0): There is no significant difference in total purchases between customers from the US and the rest of the world.
Alternate Hypothesis (H1): The US fares significantly better than the rest of the world in terms of total purchases.

The test result shows a t-statistic value of 0.3083 and a p-value of 0.7578 which is substantially higher than the typical significance level of 0.05. 
Hence, we accept the null hypothese which is that There is no significant difference in total purchases between customers from the US and the rest of the world.
